```python
A=np.array([1,-0.9])
B=np.array([1])
w, h = freqz(B, A, worN=512, whole=False)
```

Kode tersebut merupakan kelanjutan dari persiapan sebelumnya untuk melakukan analisis frekuensi menggunakan fungsi freqz dari library SciPy.

1.	A=np.array([1,-0.9]): Baris ini membuat array NumPy A yang mewakili koefisien polinomial penyebut dari sistem linier yang akan dianalisis. Dalam kasus ini, koefisien polinomial adalah 1 - 0.9z^-1.
2.	B=np.array([1]): Baris ini membuat array NumPy B yang mewakili koefisien polinomial pembilang dari sistem linier yang akan dianalisis. Dalam kasus ini, koefisien polinomial hanya memiliki satu nilai, yaitu 1.
3.	w, h = freqz(B, A, worN=512, whole=False): Baris ini memanggil fungsi freqz dari library SciPy untuk menghitung respon frekuensi dari sistem linier yang diwakili oleh koefisien polinomial B dan A. Argumen worN=512 menentukan jumlah titik frekuensi yang akan dihitung, sedangkan whole=False mengindikasikan bahwa hanya setengah dari rentang frekuensi akan dihitung (dari 0 hingga frekuensi Nyquist). Fungsi freqz mengembalikan dua array, yaitu w yang berisi nilai-nilai frekuensi angular dalam satuan radian/sampel, dan h yang berisi nilai-nilai respon frekuensi kompleks.

Kode ini melakukan perhitungan respon frekuensi untuk sistem linier dengan koefisien polinomial penyebut 1 - 0.9z^-1 dan koefisien polinomial pembilang 1. Hasil perhitungan disimpan dalam variabel w (nilai-nilai frekuensi angular) dan h (nilai-nilai respon frekuensi kompleks). Untuk memvisualisasikan hasil perhitungan, Anda perlu menambahkan kode lain yang menggunakan library Matplotlib untuk membuat plot respon frekuensi.

```python
fig = plt.figure()
plt.title('Digital filter frequency response')
plt.subplot(2,1,1)
plt.plot(w, 20 * np.log10(abs(h)), 'b')
plt.ylabel('Amplitude (dB)')
plt.xlabel('Normalized Frequency')
plt.grid()
ax = plt.gca()
ax.xaxis.set_major_formatter(FuncFormatter(
   lambda val,pos: '{:.0g}$\pi$'.format(val/np.pi) if val !=0 else '0'
))
ax.xaxis.set_major_locator(MultipleLocator(base=0.5))
plt.subplot(2,1,2)
angles = np.angle(h)
plt.plot(w, angles, 'g')
plt.ylabel('Angle (radians)')
plt.xlabel('Normalized Frequency')
ax = plt.gca()
ax.xaxis.set_major_formatter(FuncFormatter(
   lambda val,pos: '{:.0g}$\pi$'.format(val/np.pi) if val !=0 else '0'
))
ax.xaxis.set_major_locator(MultipleLocator(base=0.5))
plt.grid()
plt.tight_layout()
```

```python
def zplane(B,A):
    # Plot Poles and Zeros
    from  matplotlib import patches
    plt.figure(figsize=(6,4))
    ax = plt.subplot(111)
    r = 1.5; plt.axis('scaled'); plt.axis([-r, r, -r, r])
    #ticks = [-1, 1]; plt.xticks(ticks); plt.yticks(ticks)
    # Unit Circle
    uc = patches.Circle((0,0), radius=1, fill=False, color='black', ls='dashed')
    ax.add_patch(uc)
    ax.spines['left'].set_position('center')
    ax.spines['bottom'].set_position('center')
    ax.spines['right'].set_visible(False)
    ax.spines['top'].set_visible(False)
    plt.xlabel('Re', horizontalalignment='right', x=1.0)
    plt.ylabel('Im',  y=1.0)
    plt.title('Complex z-Plane', loc='right')
    plt.grid()
    plt.plot(np.real(A),np.imag(A),'rx')
    plt.plot(np.real(B),np.imag(B),'bo')
```

Kode tersebut adalah sebuah fungsi Python bernama zplane yang digunakan untuk memplot Pole dan Zero dari suatu sistem linier pada bidang kompleks z.

1.	def zplane(B, A): Baris ini mendeklarasikan sebuah fungsi bernama zplane yang menerima dua argumen B dan A. B adalah array yang berisi koefisien pembilang, dan A adalah array yang berisi koefisien penyebut dari sistem linier.
2.	from matplotlib import patches Baris ini mengimpor modul patches dari library Matplotlib yang digunakan untuk membuat objek-objek geometris seperti lingkaran pada plot.
3.	plt.figure(figsize=(6, 4)) Baris ini membuat sebuah figure baru dengan ukuran 6 x 4 inci.
4.	ax = plt.subplot(111) Baris ini membuat sebuah sumbu (axes) tunggal pada figure.
5.	r = 1.5; plt.axis('scaled'); plt.axis([-r, r, -r, r]) Baris-baris ini menyesuaikan batas sumbu x dan y agar lingkaran satuan dapat ditampilkan dengan baik.
6.	uc = patches.Circle((0, 0), radius=1, fill=False, color='black', ls='dashed') Baris ini membuat objek lingkaran dengan pusat di (0, 0) dan jari-jari 1, dengan garis putus-putus berwarna hitam.
7.	ax.add_patch(uc) Baris ini menambahkan lingkaran yang telah dibuat pada sumbu (axes).
8.	ax.spines['left'].set_position('center'), ax.spines['bottom'].set_position('center'), ax.spines['right'].set_visible(False), ax.spines['top'].set_visible(False) Baris-baris ini menyesuaikan tampilan sumbu agar lebih rapi dan hanya menampilkan sumbu x dan y.
9.	plt.xlabel('Re', horizontalalignment='right', x=1.0), plt.ylabel('Im', y=1.0), plt.title('Complex z-Plane', loc='right') Baris-baris ini menambahkan label sumbu x, sumbu y, dan judul plot.
10.	plt.grid() Baris ini menambahkan grid pada plot.
11.	plt.plot(np.real(A), np.imag(A), 'rx') Baris ini memplot Pole dari sistem linier yang diwakili oleh koefisien penyebut A dengan tanda silang berwarna merah.
12.	plt.plot(np.real(B), np.imag(B), 'bo') Baris ini memplot Zero dari sistem linier yang diwakili oleh koefisien pembilang B dengan tanda lingkaran berwarna biru.

Fungsi zplane digunakan untuk memvisualisasikan Pole dan Zero dari suatu sistem linier pada bidang kompleks z. Fungsi ini menerima dua argumen, yaitu B dan A, yang mewakili koefisien pembilang dan penyebut dari sistem linier. Fungsi ini membuat sebuah plot dengan lingkaran satuan, sumbu x dan y, grid, dan memplot Pole dan Zero pada bidang kompleks z menggunakan tanda silang dan lingkaran dengan warna yang berbeda.
```python
import scipy.signal
import matplotlib.pyplot as plt
#Start with a unit pulse as input x:
x=np.zeros(30)
x[0] = 1
#B and A are given as before:
A=[1,-0.9];
B=[1];
#Now calculate the impulse response:
y=scipy.signal.lfilter(B,A,x);
plt.figure()
plt.stem(y, use_line_collection=True)
plt.xlabel('Sample')
plt.ylabel('Value')
plt.grid()
```

5.	y = scipy.signal.lfilter(B, A, x): Baris ini menggunakan fungsi lfilter dari modul scipy.signal untuk menghitung respon impuls y dari sistem linier dengan koefisien B dan A serta input x.
