Example 1: Lowpass with  fs=1Hz
For this 31 tap filter we choose the cutoff frequency to be  Fc=Fs/8, or in normalized form  fc=1/8.

```python
b_k = fir_d.firwin_kaiser_lpf(1/8,1/6,50,1.0)
b_r = fir_d.fir_remez_lpf(1/8,1/6,0.2,50,1.0)
```

```python
b_k = fir_d.firwin_kaiser_lpf(1/8, 1/6, 50, 1.0)
b_r = fir_d.fir_remez_lpf(1/8, 1/6, 0.2, 50, 1.0)

print(f"Kaiser Win filter taps = {len(b_k)}.")
print(f"Remez filter taps = {len(b_r)}.")
```

Pada kode Python tersebut, terdapat dua baris kode yang menggunakan fungsi dari modul fir_d (yang sebelumnya diimpor dari sk_dsp_comm.fir_design_helper).

b_k = fir_d.firwin_kaiser_lpf(1/8, 1/6, 50, 1.0): Baris kode ini menggunakan fungsi firwin_kaiser_lpf dari modul fir_d untuk merancang filter lolos rendah (low-pass filter) FIR dengan menggunakan metode window Kaiser. Parameter yang diberikan adalah:

1.	1/8: Frekuensi normalisasi cutoff untuk filter lolos rendah.
2.	1/6: Frekuensi normalisasi transisi untuk filter lolos rendah.
3.	50: Jumlah koefisien filter yang akan dihasilkan.
4.	1.0: Nilai parameter beta untuk window Kaiser. Hasil dari fungsi ini disimpan dalam variabel b_k.

b_r = fir_d.fir_remez_lpf(1/8, 1/6, 0.2, 50, 1.0): Baris kode ini menggunakan fungsi fir_remez_lpf dari modul fir_d untuk merancang filter lolos rendah (low-pass filter) FIR dengan menggunakan metode Remez exchange. Parameter yang diberikan adalah:

1.	1/8: Frekuensi normalisasi cutoff untuk filter lolos rendah.
2.	1/6: Frekuensi normalisasi transisi untuk filter lolos rendah.
3.	0.2: Nilai ripple maksimum yang diinginkan di band lolos dan band stop.
4.	50: Jumlah koefisien filter yang akan dihasilkan.
5.	1.0: Nilai parameter beta untuk window Kaiser (digunakan dalam inisialisasi koefisien filter). Hasil dari fungsi ini disimpan dalam variabel b_r.

Kode tersebut menggunakan dua fungsi dari modul fir_d untuk merancang filter lolos rendah FIR dengan menggunakan dua metode yang berbeda: metode window Kaiser dan metode Remez exchange. Metode window Kaiser menghasilkan filter dengan karakteristik transisi yang lebih halus, sedangkan metode Remez exchange menghasilkan filter dengan ripple yang lebih rendah di band lolos dan band stop. Hasil dari kedua fungsi ini disimpan dalam variabel b_k dan b_r, yang masing-masing berisi koefisien filter yang dapat digunakan dalam operasi filtering sinyal digital. Dengan menggunakan fungsi-fungsi dari modul fir_d, pengguna dapat dengan mudah merancang filter FIR sesuai dengan spesifikasi yang diinginkan.

```python
fir_d.freqz_resp_list([b_k,b_r],[[1],[1]],'dB',fs=1)
ylim([-80,5])
title(r'Kaiser vs Equal Ripple Lowpass')
ylabel(r'Filter Gain (dB)')
xlabel(r'Frequency in kHz')
legend((r'Kaiser: %d taps' % len(b_k),r'Remez: %d taps' % len(b_r)),loc='best')
grid();
```

Pada kode Python tersebut, terdapat beberapa baris yang digunakan untuk memvisualisasikan dan membandingkan respons frekuensi dari dua filter FIR yang dirancang sebelumnya, yaitu filter Kaiser (b_k) dan filter Remez (b_r).

Parameter [b_k, b_r] adalah daftar koefisien filter, [[1], [1]] adalah daftar koefisien filter yang diasumsikan bernilai 1 (tidak ada filter IIR), 'dB' menentukan bahwa respons frekuensi akan dinyatakan dalam desibel (dB), dan fs=1 menentukan frekuensi sampel normalisasi sebesar 1.
2.	ylim([-80, 5]): Baris ini mengatur batas sumbu Y pada plot menjadi antara -80 dB dan 5 dB.
3.	title(r'Kaiser vs Equal Ripple Lowpass'): Baris ini memberikan judul pada plot sebagai "Kaiser vs Equal Ripple Lowpass".

5.	xlabel(r'Frequency in kHz'): Baris ini memberikan label pada sumbu X sebagai "Frequency in kHz".
6.	legend((r'Kaiser: %d taps' % len(b_k), r'Remez: %d taps' % len(b_r)), loc='best'): Baris ini menambahkan legenda pada plot dengan label yang menunjukkan jumlah koefisien filter untuk masing-masing filter (Kaiser dan Remez). loc='best' menentukan bahwa legenda akan ditempatkan pada posisi yang paling baik secara otomatis.

Example: Lowpass Design Comparison

The lowpass amplitude response requirements given  fs=48 kHz are:

3.	Passband ripple of 0.5 dB
4.	Stopband attenuation of 60 dB

Design four filters to meet the same requirements: butter, cheby1, ,cheby2, and ellip:

```python
fs = 48000
f_pass = 5000
f_stop = 8000
b_but,a_but,sos_but = iir_d.IIR_lpf(f_pass,f_stop,0.5,60,fs,'butter')
b_cheb1,a_cheb1,sos_cheb1 = iir_d.IIR_lpf(f_pass,f_stop,0.5,60,fs,'cheby1')
b_cheb2,a_cheb2,sos_cheb2 = iir_d.IIR_lpf(f_pass,f_stop,0.5,60,fs,'cheby2')
b_elli,a_elli,sos_elli = iir_d.IIR_lpf(f_pass,f_stop,0.5,60,fs,'ellip')
```

```python
fs = 48000
f_pass = 5000
f_stop = 8000

# Filter Butterworth
b_but, a_but, sos_but = iir_d.IIR_lpf(f_pass, f_stop, 0.5, 60, fs, 'butter')
order_but = max(len(b_but), len(a_but)) - 1
print(f"IIR butter order = {order_but}.")

# Filter Chebyshev tipe 1
b_cheb1, a_cheb1, sos_cheb1 = iir_d.IIR_lpf(f_pass, f_stop, 0.5, 60, fs, 'cheby1')
order_cheb1 = max(len(b_cheb1), len(a_cheb1)) - 1
print(f"IIR cheby1 order = {order_cheb1}.")

# Filter Chebyshev tipe 2
b_cheb2, a_cheb2, sos_cheb2 = iir_d.IIR_lpf(f_pass, f_stop, 0.5, 60, fs, 'cheby2')
order_cheb2 = max(len(b_cheb2), len(a_cheb2)) - 1
print(f"IIR cheby2 order = {order_cheb2}.")

# Filter Eliptik
b_elli, a_elli, sos_elli = iir_d.IIR_lpf(f_pass, f_stop, 0.5, 60, fs, 'ellip')
order_elli = max(len(b_elli), len(a_elli)) - 1
print(f"IIR ellip order = {order_elli}.")
```
